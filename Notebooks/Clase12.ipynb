# Programación Funcional
## Listas por comprensión 
Las listas por compresión son estructuras en las cuales podemos modificar una lista (en realidad cualquier objeto iterable) de una manera más sencilla.
### Sintáxis
`[regla_a_aplicar for elemento in lista]`
### Ejemplos

Ejemplos con listas:

```python
lista=[1,2,3]
lista=[x*x for x in lista]
print(lista)
```

```python
lista=[1,2,3]
print([x*x for x in lista])
print(lista)
```

```python
lista=[1,2,3]
lista=[x+1 for x in lista]
print(lista)
```

```python
lista=['a','b','c']
lista=[x.upper() for x in lista]
print(lista)
```

```python
lista=['Edgar','Kathy','Silvia']
lista=['Hola '+x for x in lista]
print(lista)
```

### Aplicando un criterio
Podemos aplicar ciertos *criterios* o hasta funciones a nuestras listas. Por ejemplo, vamos a imaginar que tenemos la siguiente lista:

`lista=['Omar Aparicio','Irving Suárez','Ignacio Villanueva']`

Si solamente nos interesara tener en una lista los nombres podríamos proceder de la siguiente forma

```python
lista=['Omar Aparicio','Irving Suárez','Ignacio Villanueva']
print([x.split(' ')[0] for x in lista])
```

Deja te lo traduzco.

A cada elemento de lista, el cual está compuesto de la siguiente forma: `'Nombre Apellido'`

Lo convertimos con el método `split(' ')` en una *sublista* con la forma: `['Nombre','Apellido']`

Y por último con `split(' ')[0]` nos quedamos con la primera entrada de cada *sublista*.
### Aplicando una función
Como ya te mencioné podemos aplicar una función a nivel de criterio. Veamos un ejemplo que con esos nos entendemos mejor.

```python
def saludar(nombre):
    return 'Hola '+nombre

# mi función sólo saluda a la persona que pongas como parámetro
print(saludar('Óscar'))
# implementémosla en una lista por comprensión
lista=['Edgar','Kathy','Silvia']
print([saludar(x) for x in lista])
```

### Filtrado
Por último veamos una manera de poner una condición para filtrar los datos (escoger un tipo bajo una condición) que queremos de nuestra lista.

La sintáxis es muy sencilla ya que lo único que tenemos que agregar a nuestra sintáxis original es que a la derecha ponemos la condición con la cuál haremos el filtrado:

`[regla_a_aplicar for elemento in lista if condición_de_filtrado]`

De ejemplo haremos una lista que contenga solamente los número pares del 1 al 100, dada la lista que contiene todos los números del 1 al 100.

```python
lista=list(range(1,101))
print([x for x in lista if x%2==0])
```

### Ejercicio
Graficar todos los pasos que le tomará llegar a los números del 2 al 10 para que se cumpla la Conjetura de Collatz. Usar al menos una vez las listas por comprensión. 

```python
import matplotlib.pyplot as plt

def collatz(numero):
    numeros=[]
    while True:
        numeros.append(numero)
        if numero!=1:
            if numero%2==0:
                numero/=2
            else:
                numero=3*numero+1
        else:
            break
    return numeros

lista=[collatz(x) for x in list(range(2,11))]
print(lista)
i=2
for ele in lista:
    x=list(range(1,len(ele)+1))
    plt.scatter(x,ele,color='red',linewidth=2,label='Número %s'%(i))
    i+=1
    plt.xlabel('Paso')
    plt.ylabel('Número')
    plt.legend()
    plt.grid(True)
    plt.show()
```
