# Numpy
El módulo numpy nos servirá para la implementación de arreglos y matrices. Supone una ventaja contra las listas ya que en ejecución es más rápido, al igual de que está muy relacionado como otros módulos tales como pandas. También es muy usado en inteligencia artificial.

La manera en que se hace la importación es la siguiente:

```python
import numpy as np
```

## Arrays
Los arrays son tipos de datos muy usados en todos los lenguajes de programación y nosotros los podemos pensar como una caja que tiene elementos del mismo tipo. 
### Crear arrays dada una lista
Podemos crear arrays de diferentes formas. La primera que veremos será dada una lista y usaremos el método *array()*

```python
lista=[1,2,3,4]
arreglo=np.array(lista)
print(arreglo)
```

Podemos saber qué tipo de datos contiene el arreglo con el método *dtype*

```python
arreglo.dtype
```

Y darle un tipo de dato al momento de crearlos

```python
arreglo=np.array(lista,dtype='int16')
arreglo.dtype
```

### Creando arrays con el método arange()
Con el método *arange(inicio,final,incremento)* nos permite contruir un arreglo con los números del *inicio* al *final-1* y con incrementos según le pongamos (si no se pone el incremento será de uno en uno).

```python
a=np.arange(0,5)
print(a)
b=np.arange(0,5,2)
print(b)
```

## Matrices
Para crear matrices el secreto será pensar que una matriz es una arreglo donde cada elemento es un array. 

```python
matriz=np.array([[1,2,3],[4,5,6]])
matriz
```

## Matrices especiales
En el álgebra lineal tenemos ciertas matrices (y vectores) que son muy particulares, tales como la identidad. Vamos a ver que hay maneras en las que nosotros podremos crear estas de manera automática.
### Matriz identidad
Usamos el método *eye(n)* donde *n* será el tamaño de la matriz identidad.

```python
identidad=np.eye(4)
print(identidad)
```

### Arreglo de ceros
Usaremos el método *zeros(n)* con *n* el tamaño de nuestro arreglo.

```python
ceros=np.zeros(4)
print(ceros)
```

### Matriz de ceros
Implementamos el método *zeros((renglón,columna))*.

```python
ceros=np.zeros((2,5))
print(ceros)
```

### Arreglo de unos
Será lo mismo que para el de ceros.

```python
unos=np.ones(3)
print(unos)
```

### Matriz de unos

```python
unos=np.ones((3,4))
print(unos)
```

## Acceder a valores
Procederemos usando los índices y el slicing.
### A una columna

```python
a=np.eye(5)
print(a)
print(a[0:3,3]) # tomamos toda la columna 3
```

### A un renglón

```python
a=np.eye(5)
print(a)
print(a[3,2:]) # tomamos todo el tercer renglón
```

### A un elemento en específico

```python
a=np.array([[0,1,2],[3,4,5]])
print(a)
print(a[0,1]) # renglón 0 columna 1
```

### A varios elementos de una columna

```python
a=np.array([[0,1,2,12],[3,4,5,12],[6,7,8,12],[9,10,11,12]])
print(a)
print(a[0:2,1]) # de la columna 1 a los elementos 0 y 1
```

### A varios elementos de un renglón

```python
a=np.array([[0,1,2,12],[3,4,5,12],[6,7,8,12],[9,10,11,12]])
print(a)
print(a[3,1:]) # del renglón 3 del uno en adelante
```

### Otra forma 

```python
a=np.arange(10)
# accedemos a los elementos del 4 en adelante saltando de 2 en 2
print(a[4::2])
```

## Modificar elementos
Para modificar elementos vamos a hacerlo mediante asignación.
### Modificar un elemento de un array

```python
a=np.arange(10)
a[4]=1000
print(a)
```

```python
list(range(9))
```

### Modificar un elemento de una matriz

```python
a=np.array([list(range(5)),list(range(5,10))])
print(a)
a[0,-1]=1000 # renglón 0, columna 4
print(a) 
```

## Multiplicación de arrays
Es igual a la multiplicación vectorial.

```python
a=np.arange(5)
b=np.arange(5,10)
print(a,b)
a*b
```

## Traspuesta de una Matriz
Para obtener la matriz traspuesta solamente hay que usar el atributo T

```python
b=np.array([list(range(4,6)),list(range(6,8))])
print(b,b.T,sep="\n")
```

## Multiplicación de Matrices
La hacemos con el método *dot()*

```python
a=np.array([list(range(2)),list(range(2,4))])
b=np.array([list(range(4,6)),list(range(6,8))])
print(a,b,sep="\n")
print(a.dot(b))
```

## Dimensiones de un matriz
Con el atributo *shape()* te da las dimensiones en una tupla de la forma (renglón,columna)

```python
a=np.array([list(range(2)),list(range(2,4)),list(range(2,4))])
print(a)
a.shape
```

## Agregar elementos a un array
Utilizamos el método *append()*

```python
a=np.array(list(range(3)))
np.append(a,1000)
```

## Insertar una columna
Tanto para insertar renglones o columnas usamos la sitáxis: *numpy.insert(arr, obj, values, axis=None)*

```python
a=np.array([list(range(2)),list(range(2,4)),list(range(2,4))])
np.insert(a, a.shape[1], np.array((10, 10, 10)), 1)
```

**Nota**: Usamos el atributo shape ya que es una tupla con los reglones y columnas.
## Insertar un renglón

```python
a=np.array([list(range(2)),list(range(2,4)),list(range(2,4))])
np.insert(a, a.shape[0], np.array((10, 10)), 0)
```
